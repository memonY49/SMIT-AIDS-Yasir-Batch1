# question1
- I first year the investment was of 2500$ where the profit on the investment at the and of the year 1 is 25%
- For the sencond year it is 50%
- For the third year it is 100%
- For the forth year it is -50%
- and in last year 25%
- find out the a-mean,g-mean and the total amount in my bank.
- 

```python
import math
in_invest =2500
roc = np.array([0.25,0.5,1,-0.5,0.25])
# roc = np.array([1,-0.5])
factor =  roc + 1
product = np.prod(factor)
print(product)
gmean = math.pow(product,1/len(roc))-1
print(f"Initial Investment:{in_invest}$")
print(f"Amount in my Bank:{in_invest * product}$")
print(f"AMean:{round(np.mean(roc)*100,2)}%")
print(f"GMean:{round(gmean*100,2)}%")
```

# Tomorrow's task is to analyze the attached csv file which contain columns (Year, Profit, additional investment and total amount of every year) remember the initial investment is of 5000$.

Output:
- Initial Investment: ??
- Total Profit Per Interval of Every year (GMean): ??
- Arithmetic Mean: ??
- Total Amount in my Bank: ??
- Total Investment: ??
- List of loss making years: ??
- List of Profit making years: ??

```python
my_file = np.loadtxt("InvestmentDataset.csv", delimiter = ',',dtype = str)
features = my_file[0]
in_invest = 5000
print(features)
my_file = np.delete(my_file,0, axis=0) #remove headers
roc = np.array([pro.replace("%","") for pro in my_file[:,1]]).astype(float)
add_invest = my_file[:,2].astype(float)
loss_years = np.where(roc<0)
profit_years = np.where(roc>0)
amean = np.mean(roc/100)
roc = (roc / 100)+1
product = np.prod(roc)
gmean = math.pow(product,1/len(roc)) - 1
iloc_AI = np.where(add_invest != 0)
print(my_file[iloc_AI[0],3].astype(float)-add_invest[iloc_AI[0]])
print(f"GMean: {gmean*100}")
print(f"AMean: {amean}")
print(f"Total amont in bank: {my_file[-1,-1]}")
print(f"Total investment: {in_invest + np.sum(add_invest)}")
print(f"Loss making Years: {my_file[loss_years[0],0]+' Year'}")
print(f"Profit making Years: {my_file[profit_years[0],0]+' Year'}")
```

```python
import pandas as pd
import numpy as np

np.random.seed(0)

# Generate customer IDs
customer_ids = [f"CID-{str(i).zfill(3)}" for i in range(1, 11)]

# Generate order dates
order_dates = pd.date_range(start="2022-01-01", periods=100).strftime("%Y-%m-%d")

# Generate totals
totals = np.random.uniform(10, 100, size=100).round(2)

# Generate customer IDs for each order
order_customer_ids = np.random.choice(customer_ids, size=100)

# Create a DataFrame
df = pd.DataFrame({
    'Order ID': range(1, 101),
    'Customer ID': order_customer_ids,
    'Order Date': order_dates,
    'Total': totals
})

# Save to CSV file
df.to_csv('orders.csv', index=False)
```
