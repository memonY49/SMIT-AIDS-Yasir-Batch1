# Task Analysis of student data
- add a column of Total marks and obtain marks
- add a column of %
- add a column of grade
- add a column of class
- find how many of students have (A,B,C,D or fail) Grades in every class
- find how many of students have (A,B,C,D or fail) Grades in total
- find max marks student in every class
- find min in every class
- find max in total
- find min in total 

```python
students_data['total'] = 100
students_data['obtain'] = students_data[['exam1','exam2','final_exam']].sum(axis = 1)
```

```python
students_data['percentage'] = students_data['obtain']/students_data['total']*100
```

```python
def grades(score):
    if score >= 80:
        return 'A' 
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'
        
students_data['grades'] = students_data['percentage'].apply(grades)
```

```python
students_data.drop('class', axis = 1, inplace = True)
```

```python
students_data['class'] = students_data.roll_no.str.split('/').str[1]
```
