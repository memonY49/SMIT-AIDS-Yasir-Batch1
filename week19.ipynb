```python
students_data = pd.read_csv("students_five_classes.csv")
```

```python
students_data.head()
```

```python
students_data['class'] = students_data['roll_no'].str.split('/').str[:2].str.join('/')
```

```python
students_data['total'] = 100
students_data['obtain'] = students_data[['exam1','exam2','final_exam']].sum(axis = 1)
```

```python
students_data['percentage'] = students_data['obtain']/students_data['total']*100
```

```python
def grades(score):
    if score >= 80:
        return 'A' 
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'
        
students_data['grades'] = students_data['percentage'].apply(grades)
```

```python
students_data.groupby('class').get_group('SMIT/DSAI').grades.value_counts()
```

```python
student_pivot = students_data.pivot_table(index = 'grades', columns = 'class', aggfunc = 'size', fill_value = 0).sort_index()
student_pivot
```

```python
grouped = students_data.groupby('class')

for st_class in students_data['class'].unique():
    chart_value = grouped.get_group(st_class).grades.value_counts()
    plt.title(f'Class: {st_class}')
    plt.xlabel('Grades')
    plt.ylabel('No of Students')
    plt.bar(chart_value.index,chart_value.values)
    plt.show()
```
